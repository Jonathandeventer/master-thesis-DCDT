```
#rdyn_instance= "2000_100_15_0.6_0.8_0.2_1_2.5_0"
import os
import time
import pandas as pd

from nf1 import NF1
#results_fn = listdir van rdyn_instances


res_fn = os.listdir(datapath+data_folder)

def read_coms(filename):
        com = []
        with open(filename) as f:
            for l in f:
                try:
                    l = l.replace("(","[").replace(")","]").replace("]", "").replace(" ", "").split("[")[1]
                except:
                    pass
                ns = tuple(map(int, l.rstrip().replace(" ", "\t").replace(",", "\t").split("\t")))

                if len(ns) == 2:
                    raise Exception
                com.append(ns)
        return com

rows_list= []
for _, rdyn_instance in enumerate(res_fn):  
    print(_, rdyn_instance)
    # get gt comms
    fn = os.listdir(datapath+data_folder+rdyn_instance)
    comms = [x for x in fn if 'communities' in x] 
    #print(comms)
    #print(comms)

    comm_nr = []
    for c in comms:
        comm_nr.append(c[12:].replace(".txt",''),)

    # get identified comms
    i_comms = os.listdir(output_folder + rdyn_instance)
    i_comms_i = [x for x in i_comms if "PostMerg" in x]

    #print(i_comms_i)

    
    count = 0
    for c, v in enumerate(i_comms_i):
        #print(c,v)
        if str(c) in comm_nr:

            identified_coms = []
            #print("11")
            with open(output_folder + rdyn_instance + "/"+v) as f:
                com = []
                for l in f:
                    line = l.split(':')
                    com = line[1].split()
                    commm = [int(x) for x in com]
                    identified_coms.append(tuple(commm))

            gt = read_coms(datapath+data_folder+rdyn_instance+'/'+comms[count])
            count +=1
            #print(c,datapath+data_folder+rdyn_instance+'/'+comms[c])
            if identified_coms == []:
                    continue
            #print(output_folder + rdyn_instance + "/"+v)
            #print(comm_nr)
            #print(comms)

            nf = NF1(identified_coms, gt)
            #print(nf.summary())
            res = []
            res.extend(nf.get_f1())
            #print(nf.get_partition_stats())
            #print ()
            mean = str(res[0])
            std = str(res[1])
            maxx = str(res[2])
            minn = str(res[3])
            mode =str(res[4])


            row = {
                "algorithm": algorithm_name,
                "rdyn_instance": str(rdyn_instance),
                #"graph_n": str(alg_results_fn[i]),
                "mean": mean.replace(".",","),
                "std" : std.replace(".",","),
                "min": minn.replace(".",","),
                "max": maxx.replace(".",","),
                "mode": mode.replace(".",",")
            }
            rows_list.append(row)
        #print(rows_list)
        #print(os.getcwd())
        #else:
            #print('skip')

df = pd.DataFrame(rows_list)           
df.to_csv(datapath+"/output_t_"+algorithm_name+".txt", sep=' ', index=False)



```

```
i = "/interactions.txt"
import pandas as pd
import os
import shutil
import time
os.chdir("/content")
#print(os.getcwd())

results_fn = os.listdir(datapath+data_folder)
print(len(results_fn))
times= []

for _, rdyn_instance in enumerate(results_fn):
    
    print(_," ",os.getcwd())
    #break

    # extract, load and transform data in algorithm.
    data = pd.read_csv(datapath+data_folder+rdyn_instance + i, sep="\t", header=None)
    data.columns = ["iteration", "timestamp", "action", "node_id1", "node_id2"]
    data['node_id1'] = data['node_id1'].astype(str)
    data["placeholder"] = " "
    data['node_id2'] = data['node_id2'].astype(str)
    data['edge'] = data[['node_id1','placeholder', 'node_id2']].apply(lambda x: ''.join(x), axis=1)
    data['edge2'] = data[['node_id2','placeholder', 'node_id1']].apply(lambda x: ''.join(x), axis=1)

    count = 0
    res = []

    nodes_0 = set()
    edges_0 = list()
    changes = {}
    for el in range(len(data["edge"])):
        #print (data["iteration"][el], data["edge"][el], data["edge2"][el])

        if data["iteration"][el] == 0:
            nodes_0.add(data["node_id1"][el])
            nodes_0.add(data["node_id2"][el])
            edges_0.append(data["edge"][el])
            edges_0.append(data["edge2"][el])

        else:
            if data["iteration"][el] not in changes.keys():
                changes[data["iteration"][el]] = []
                changes[data["iteration"][el]].append("{0} {1}".format(data["action"][el], data["edge"][el]))
            else:
                changes[data["iteration"][el]].append("{0} {1}".format(data["action"][el], data["edge"][el]))
           
    os.mkdir(datapath + 'hoctracker/'+rdyn_instance)
    src = 'gdrive/My Drive/data/hoctracker/HOCTracker.jar'
    dst = datapath + 'hoctracker/'+rdyn_instance + '/HOCTracker.jar'
    shutil.copy(src, dst)
    
    
    with open(datapath + 'hoctracker/'+rdyn_instance + "/network.net", 'w') as f:
        size = len(nodes_0)
        f.write("*Vertices {0}\n".format(size))
        for el in nodes_0:
            f.write("{0} {1}\n".format(el,el))

        f.write("*Edges {}\n".format(2*len(edges_0)))

        for ed in edges_0:
            f.write("{} {}\n".format(ed,ed,1))


    for key, value in changes.items():
        #print (key, value)
        #break
        with open(datapath + 'hoctracker/'+rdyn_instance + "/change"+str(key)+".dat", 'w') as f:
            for ed in value:
                if ed[0] == "+":
                    f.write("{}\n".format(ed[2:]))
                elif ed[0] == "-":
                    f.write("{}\n".format(ed[0]+ed[2:]))

     # At this point the necessary files for one rdyn instance are made.
    
     # move jar in map and run algorithm.
    
    os.chdir(datapath + '/hoctracker/'+rdyn_instance)
    print(os.getcwd())
    start = time.time()
    !java -jar HOCTracker.jar -f network.net -t 100 > /dev/null
    end = time.time()
    x = str(end-start)
        
    run_time = {   "algorithm": algorithm_name,
                   "rdyn_instance_name" : rdyn_instance,
                   "execution_time" : x.replace(".",",") 
    }
    times.append(run_time)
    os.chdir("/content")
    
    
df = pd.DataFrame(times)  
df.to_csv(datapath+"execution_times_t1_"+algorithm_name+".txt", sep=' ', index=False)
```
